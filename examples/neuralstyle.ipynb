```python
import sys
sys.path.insert(0, '..') # add root to python path

import eden.setup
from eden.utils import dataset
from eden.utils import processing
from eden.canvas import canvas
from eden.external import neuralstyle

eden.setup.set_external_repos_root("../external")

data = dataset.ImageLoader()
img1 = data.load_image('../images/dog.jpg')
img2 = data.load_image('../images/starrynight.jpg')

params = {
    'content': img1['data'],
    'style': img2['data'],
    'image_size': 1024,
    'num_iterations': 1000
}

img_out = neuralstyle.run(params)
processing.show(img_out)
```
